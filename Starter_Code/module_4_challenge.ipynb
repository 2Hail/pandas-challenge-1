## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
data["subtotal_calculation"] = data["unit_price"] * data["qty"]
price_quantity_subtotal = data[["unit_price", "qty", "subtotal_calculation"]]
price_quantity_subtotal.head(2)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
def shipping_price(unit_weight):
    if unit_weight > 50:
        return 7 * unit_weight
    else:
        return 10 * unit_weight
    
data["total_weight"] = data["unit_weight"] * data["qty"]

data["shipping_price"] = data["total_weight"].apply(shipping_price) 
data["shipping_price"] = data["shipping_price"].round(2)

shippingtotal = data[["unit_price", "unit_weight", "qty", "total_weight", "shipping_price"]]

shippingtotal.head(3)
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
data["line_price"] = data["shipping_price"] + data["subtotal_calculation"] 

sales_tax = 0.0925

data["line_price"] = data["line_price"] * (1 + sales_tax)
data["line_price"] = data["line_price"].round(2)

data.head(3)
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
data["line_cost"] = (data["unit_cost"] * data["qty"]) + data["shipping_price"]
data["line_cost"] = data["line_cost"].round(2)

data.head(3)
```

```python
# Create a column for the profit of
# each line using line cost and line price
data["line_profit"] = data["line_price"] - data["line_cost"]
data["line_profit"] = data["line_profit"].round(2)

data.head(3)
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.
```

```python
top_5_client_entries = data[data["client_id"].isin(top_5_client_ids)].copy()
summary_df = top_5_client_entries.groupby("client_id").agg(
    qty=("qty","sum"),
    shipping_price=("shipping_price", "sum"),
    line_price=("line_price", "sum"),
    line_cost=("line_cost", "sum")
)
summary_df["Total Revenue Holder"] = summary_df["line_price"]
summary_df["Total Cost Holder"] = summary_df["line_cost"]
summary_df["Total Profit Holder"] = summary_df["line_price"] - summary_df["line_cost"]
summary_df["Shipping"] = summary_df["shipping_price"]

summary_df.reset_index(inplace=True)

summary_df["Total Revenue"] = (summary_df["Total Revenue Holder"] / 1000000).map("${:.2f}M".format)
summary_df["Total Cost"] = (summary_df["Total Cost Holder"] / 1000000).map("${:.2f}M".format)
summary_df["Total Profit"] = (summary_df["Total Profit Holder"] / 1000000).map("${:.2f}M".format)
summary_df["Shipping"] = (summary_df["shipping_price"] / 1000000).map("${:.2f}M".format)

summary_df.rename(columns={
    "client_id": "Client ID",
    "qty": "Units",
    "shipping_price": "Shipping",
    "line_price": "Total Revenue",
    "line_cost": "Total Cost",
    "line_profit": "Total Profit"
}, inplace=True)

summary_df = summary_df.sort_values(by="Total Profit", ascending=False)

summary_df.drop(columns=["Total Revenue Holder", "Total Cost Holder", "Total Profit Holder"], inplace=True)

summary_df.head()
```
